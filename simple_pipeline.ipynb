```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(42)
X = np.linspace(0,1,100)
df = pd.DataFrame(X, columns=["X"])
y = 23 * X + 1 + np.random.randn(1, 100)
df["Y"] = y[0]
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,5))
ax.scatter(df["X"], df["Y"])
ax.set(title='Data distribution', xlabel='X', ylabel='Y')
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df["X"], df["Y"], test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(np.array(X_train).reshape(-1,1), np.array(y_train))
y_pred = model.predict(np.array(X_test).reshape(-1,1))
```

```python
print(model.coef_,model.intercept_)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('Mean Squared Error:', round(mse,2))
print('R^2 Score:', round(r2,2))
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,5))
ax.scatter(X_test, y_test)
ax.scatter(X_test, y_pred)
ax.set(title='ACT vs PRED', xlabel='X', ylabel='Y')
ax.legend(['Actual values', 'Predictions'])
plt.show()
```
